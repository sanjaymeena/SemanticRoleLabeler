```python

data_path="data/srl_test.json"

with open(data_path) as json_data:
    data = json.load(json_data)

data=data['sentences']    
df=pd.DataFrame(data)
```

```python
df.head()
```

```python
# preds=df['predicates'];
# preds.columns=['predicates']
# preds.head()
df['predicates'] = df['predicates'].apply(tuple)
df.head()
df_tokens_temp=df['predicates'].value_counts(dropna=True)
```

```python
f = plt.figure(figsize=(45, 75)) # Change the size as necessary
#Plot the data:
df_tokens_temp.plot.barh(ax=f.gca(),color=my_colors) # figure.gca means "get current axis"
```

```python
df1 = df['src']
df1.value_counts(dropna=True)
```
